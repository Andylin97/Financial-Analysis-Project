```python
import numpy as np
```

```python
x=np.linspace(0,5,11) #linespace function就是在x-y这个range中平均取11个数，每个数之间的差都一样
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional Method

plt.plot(x,y)
plt.xlabel('X Label') #设置x坐标轴名字的方法
plt.ylabel('Y Label')  #设置y坐标轴名字的方法
plt.title('Title')    #设置图标名字的方法
#plt.show()
```

```python
plt.subplot(1,2,1) #subplot是生成小图表的方式。
plt.plot(x,y,'r')  #'r'代表红色图表
#plt.subplot(number of row希望几行，number of colume希望几列，这是第几个plot)
plt.subplot(1,2,2)
plt.plot(y,x,'b') #'b'代表蓝色图表
```

```python
#Object Orited Method

figure=plt.figure()
axes=figure.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
#axes.plot(y,x)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
figure=plt.figure()

axes1=figure.add_axes([0.1,0.1,0.8,0.8])
axes2=figure.add_axes([0.5,0.1,0.4,0.2])
#figure.add_axes([距离图纸左边的距离，距离图纸底边的距离，图表的长度，图表的宽度])。所有这些都不能超过1，1代表100%图纸
axes1.plot(x,y)
axes1.set_title('Big Plot')

axes2.plot(y,x)
axes2.set_title('Small Plot')
```

```python
figure=plt.figure() #先create一个empty figure
axis=figure.add_axes([0,0,1,1])  #添加一个图表,然后输入图表的大小信息,必须([因为是list信息，所以必须有中括号里面])
axis.plot(x,y) #然后.plot在图表中操作
```

```python
#Matplotlib Part Two
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2) #画subplots的方法
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout() #这一行代码的目的就是让每个表格之间的距离稍微大点，防止字重叠看不清
```

```python
#Figure Size and DPI(Dot per Inch每英寸点数，DPI越高图片精度越高)
fig=plt.figure(figsize=(3,2)） #3代表长，2代表宽
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(5,3))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#保存python return出来的figure到jpg或者png格式
fig.savefig('my_picture.jpg',dpi=200)
#fig.savefig('my_picture.png')
```

```python
figure=plt.figure()
axes=figure.add_axes([0,0,1,1])
axes.plot(x,x**2,label='X Squared') #只有给线设置label才能使用legend公式来区分线
axes.plot(x,x**3,label='X Cubed')

#axes.legend() #legend function是给每一行线加入名称的代码，前提是在plot划线是就要
axes.legend(loc=0) #这边的loc是指，希望线条的颜色说明在图的哪个位置。有专门的的词典，0代表系统自动选择的最优位置
```

```python
# Matplotlib Part Three
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=2,alpha=0.5,linestyle=':',marker='o',markersize=10, markerfacecolor='yellow',markeredgewidth=3, markeredgecolor='green') #给线设置颜色 #Linewidth是用来设计线条粗细的 #alpha是线条的透明程度（深度）alpha越小线条越透明，alpha越大线条越不透明
#linestyle是设计线条风格，可以‘---'也可以':'(:就是点状图)，还可以'steps'(阶梯图)
#marker='o'是给线上指定的数值加上marker
#markersize 是用来改变marker size的
#markerfacecolor 是设置marker颜色
#markeredgewidth 是用来设置marker边框粗细的
#markeredgecolor 是用来设置marker边框颜色的
ax.set_xlim([0,1]) #set_xlim就是只看x在这个range中的图
ax.set_ylim([0,2]) #set_ylim就是只看y在这个range中的图
```

```python
idx 
```

```python
idx1=mcdon.loc['2007-01-01':'2007-05-01'].index
idx1
#只读取2007年1月1日到5月1日之间的date数据
```

```python
stock=mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
stock
#将2007年1月1日至2007年5月1日间的stock closing value提取出来
```

```python
fig,ax=plt.subplots()
ax.plot_date(idx1,stock,'-')

#ax.yaxis.grid(True) #在y轴上加上横线，辅助阅读
#ax.xaxis.grid(True) #在x轴上加上横线，辅助阅读

fig.autofmt_xdate() #自动format x轴date的format
plt.tight_layout() #有的时候没那么好用
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(idx1,stock)
```

```python
fig,ax=plt.subplots()
ax.plot_date(idx1,stock,'-')

#ax.yaxis.grid(True) #在y轴上加上横线，辅助阅读
#ax.xaxis.grid(True) #在x轴上加上横线，辅助阅读

#Locating
ax.xaxis.set_major_locator(dates.MonthLocator()) #Format date的格式只精确到月份
#Formating
ax.xaxis.set_major_formatter(dates.DateFormatter('\n%b-%y')) #改变date的格式

ax.xaxis.set_minor_locator(dates.WeekdayLocator())
#设置minor locator，显示weekday
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
#设置minor format,如果没有这行minor format无法显示出来

fig.autofmt_xdate() #自动format x轴date的format
plt.tight_layout() #有的时候没那么好用
```
